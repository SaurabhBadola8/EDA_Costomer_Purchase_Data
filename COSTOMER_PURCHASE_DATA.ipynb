```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df=pd.read_csv("CustomerPurchaseData.csv")
df.head()
```

#### Now there is nothing missing value

```python
df.describe()
```

### Z_CostContact and Z_Revenue have constant value for all costomer data

```python
df=df.drop(columns=["Z_CostContact", "Z_Revenue"],axis=1)
df.head()
```

### Correlation coefficients greater than 0.5 are considered high, while coefficients between 0.3 and 0.5 are considered medium.

```python
correlation_matrix = df.corr()

plt.figure(figsize=(30,20 ))
sns.heatmap(correlation_matrix, annot=True, vmin=0.3, vmax=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Check which income category Response most using hisplot

```python
with_response = df[df['Response'] > 0]
without_response = df[df['Response'] == 0]
plt.figure(figsize=(10, 6))
sns.histplot(with_response['Income'], color='black', label='With Response')
sns.histplot(without_response['Income'], color='red', label='Without Response')
plt.title('Distribution of Income by Campaign Response')
plt.xlabel('Income')
plt.legend()
plt.show()
```

### combine some columns and create new column

(Kidhome , Teenhome) - Kids

(MntWines,MntFruits,MntMeatProducts,MntFishProducts,MntSweetProducts,MntSnacksProds) - Expenses

NumDealsPurchases,NumWebPurchases,NumCatalogPurchases,NumStorePurchases,NumWebVisitsMont) - NumTotalPurchase



and the old column


```python
df['Kids'] = df['Kidhome'] + df['Teenhome']
df['Expenses'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntSnacksProds']
df['TotalAcceptedCmp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response']
df['NumTotalPurchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'] + df['NumDealsPurchases']
```

```python
col_del=["AcceptedCmp1" , "AcceptedCmp2", "AcceptedCmp3" , "AcceptedCmp4","AcceptedCmp5", "Response","NumWebVisitsMonth", "NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases" , "Kidhome", "Teenhome","MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntSnacksProds"]
df=df.drop(columns=col_del,axis=1)
df.head()
```

### in the Education grouping 
(Phd,2n cycle, graduation, Master) - PG

and

(Basic) - UG

df['Education'] = df['Education'].replace(['PhD','Cycle','Graduation', 'Master'],'PG')  
df['Education'] = df['Education'].replace(['Basic'], 'UG')

```python
education_counts = df['Education'].value_counts()
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.show()
```
